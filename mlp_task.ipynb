```python
import numpy as np
```

```python
# Define the sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Define the derivative of sigmoid for future use (not required for forward pass)
def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
# Define the MLP architecture
input_size = 5   # Number of input features 
hidden_size = 4  # Number of hidden units
output_size = 1  # Number of output units
```

```python
# Initialises random weights and baises
# Weights between inputs and hidden layer
W1 = np.random.rand(input_size, hidden_size)

# Biases for hidden layer
b1 = np.random.rand(hidden_size)

# Weights between hidden and output layer
W2 = np.random.rand(hidden_size, output_size)

# Biases for output layer
b2 = np.random.rand(output_size)
```

```python
# Performing forward propagation 
def forward_pass(inputs):
    # Hidden layer computation
    hidden_layer_input = sigmoid(np.dot(inputs, W1) + b1)
    hidden_layer_output = sigmoid(np.dot(hidden_layer_input, W2) + b2)
    return hidden_layer_output * 10  # Scale output to 0-10 range
```

```python
# Sample input data (5 features - number of inputs, integers 1-100)
sample_input = np.array([[50, 80, 70, 60, 90]])
```

```python
# Validate functionality by passing sample input
output = forward_pass(sample_input)
print('Quality rating (1-10):', output)
```

```python
# Example of multiple inputs 
sample_inputs = np.array([[50, 80, 70, 60, 90],
                          [30, 40, 50, 60, 70],
                          [90, 85, 80, 75, 95]])
outputs = forward_pass(sample_inputs)
print('\nMultiple quality ratings: ')
print(outputs)
```
